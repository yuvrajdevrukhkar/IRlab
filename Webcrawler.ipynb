```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the "Books to Read" page on Goodreads
url = "https://www.goodreads.com/list/show/264.Books_to_Read"

try:
    # Send an HTTP GET request to the URL
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract book titles
        book_titles = []

        # Find the HTML structure for book titles on the Goodreads page
        title_elements = soup.select('a.bookTitle span')
        for title in title_elements:
            book_titles.append(title.text.strip())

        # Print the extracted book titles
        for i, title in enumerate(book_titles):
            print(f"Book {i + 1}: {title}")

    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```
